## Standarize column names
- Change column names to lower case
- Remove blank spaces and replace them with '_'

```python
def standarize_columns (df):
    '''
This functions replaces the names of the columns by the lower case version,
blank spaces for underscores '_'
and rename column of states
input: data frame
    '''
    cols2 = []
    #df2 = df.copy()
    for i in range(len(df.columns)):
        cols2.append(df.columns[i].lower())
    df.columns = cols2
    df.columns = df.columns.str.replace(' ', '_')
    df = df.rename(columns={'st':'state'}) # for some reason this change is not permanently applied
    
   
```

```python
standarize_columns(costumer1)
standarize_columns(costumer2)
standarize_columns(costumer3)
```

## Correct column 'costumer lifetime value'
Correct the values in the column 'customer lifetime value'. They are given as a percent, so divide them by 100 and change dtype to numerical type. (only 2 of the files need this step.)

```python
print("Costumer analysis file 1 (df:'costumer1_drop)")
display(costumer1_drop.head())
print("Costumer analysis file 2 (df:'costumer2)")
display(costumer2.head())
print("Costumer analysis file 3 (df:'costumer3)")
display(costumer3.head())
```

```python
cos1_rem = costumer1_drop.copy()
cos1_rem['customer_lifetime_value'] = cos1_rem['customer_lifetime_value'].str.rstrip('%') # Remove the '%' symbol before transforming the column dtype and performing operations
cos1_rem['customer_lifetime_value'] =  pd.to_numeric(cos1_rem['customer_lifetime_value'], errors='coerce')
cos1_rem['customer_lifetime_value']=round(cos1_rem['customer_lifetime_value'], -1)/100
display(cos1_rem.head())
```

```python
cos2_rem = costumer2.copy()
cos2_rem['customer_lifetime_value'] = cos2_rem['customer_lifetime_value'].str.rstrip('%') # Remove the '%' symbol before transforming the column dtype and performing operations
cos2_rem['customer_lifetime_value'] =  pd.to_numeric(cos2_rem['customer_lifetime_value'], errors='coerce')
cos2_rem['customer_lifetime_value']=round(cos2_rem['customer_lifetime_value'], -1)/100
display(cos2_rem.head())
```

```python
cos3_rem = costumer3.copy()
cos3_rem['customer_lifetime_value'] = cos3_rem['customer_lifetime_value'].str.rstrip('%') # Remove the '%' symbol before transforming the column dtype and performing operations
cos3_rem['customer_lifetime_value'] =  pd.to_numeric(cos3_rem['customer_lifetime_value'], errors='coerce')
cos3_rem['customer_lifetime_value']=round(cos3_rem['customer_lifetime_value'], -1)/100
display(cos3_rem.head())
```

## Concatenate the 3 data frames

```python
cost_analysis = pd.concat([cos1_rem, cos2_rem, cos3_rem], axis=0)
print("The number of rows and columns of the data frame is",cost_analysis.shape)
cost_analysis = cost_analysis.reset_index(drop=True)
display(cost_analysis)
```
